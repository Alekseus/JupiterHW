У кубической функции нет как таковой вершины.

Чтобы найти экстремумы функции нужно решить уравнение 3x^2 - 10x -30 = 0

```python

x = Symbol('x')
print(solve(3*x**2 - 10*x - 30, x))
```

Функция возрастает на промежутках от (-∞;5/3 - sqrt(115)/3] U [5/3 - sqrt(115)/3; +∞)

Функция убывает на промежутках от [5/3 - sqrt(115)/3; 5/3 + sqrt(115)/3]

f > 0 на промежутках от (-3, 4 - sqrt(22)) U (4 + sqrt(22), +∞)
